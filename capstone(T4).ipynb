```python
def fill_missing_age(a):
    median_age = a.groupby('Rex in Yrs')['Age'].transform("median")
    a['Age'] = a['Age'].fillna(median_age)
    return a

df_res = fill_missing_age(res2)
print(df_res)
```

```python
def fill_missing_Offeredband(a):
    mode_ob = a.groupby('Rex in Yrs')['Offered band'].agg(lambda x: x.mode().iloc[0] if not x.mode().empty else None)
    a['Offered band'] = a['Offered band'].fillna(a['Rex in Yrs'].map(mode_ob))
    return a

df_res2 = fill_missing_Offeredband(res2)
```
